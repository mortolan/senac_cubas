### Exercise 1 - Find the most popular movie

The zip file file "ml-100k" contains informations about 100k movies ratings. Let's find out which is the most popular movie!

To remember the data structure, import the file "u.data" in a spark's RDD and check the first 5 elements

```python
rdd1 = sc.textFile('ml-100k/u.data')
rdd2 = sc.textFile('ml-100k/u.item')

rdd1.take(5)
```

Each sample is a review of a movie. Use a Spark Job to print the movie_id and how many times this movie appears (do it in a sorted way).

(Remember, the columns are **user_id**, **movie_id**, **rating value** and **time stamp**)

```python
results = rdd1.map(lambda x: x.split()[1])\
.map(lambda x: (x, 1))\
.reduceByKey(lambda x, y: x + y)\
.sortBy(lambda x: x[1], ascending=False)\
.collect()

for i in results:
    print('{} {}'.format(i[0], i[1]))
```

What is the most popular movie?

```python
results = rdd2.map(lambda x: x.split()[0])\
.map(lambda x: (x, 1))\
.reduceByKey(lambda x, y: x + y)\
.sortBy(lambda x: x[1], ascending=False)\
.collect()

for i in results:
    print('{} {}'.format(i[0], i[1]))
```

```python
def loadMovieNames():
    movieNames = {}
    with open ("ml-100k/u.item",errors='ignore') as f:
        for line in f:
            fields = line.split('|')
            movieNames[int(fields[0])] = fields[1]
        return movieNames    
```
