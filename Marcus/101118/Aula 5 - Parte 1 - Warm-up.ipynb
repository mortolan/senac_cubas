### Exercise 1

The Book.txt file contains the content of a book called "Self-Employment: Building an Internet Business of One
Achieving Financial and Personal Freedom through a Lifestyle Technology Business". Let's find some information about it!

Load the dataset into a spark RDD and check the first 10 samples.

```python
sc = SparkContext()
```

```python
se = sc.textFile('Book.txt')
se.take(10)
```

What is each rdd sample?

Each sample is a book line

Using some Spark Job (Transformation(s) + Action) to check the number of times that each word appears in the book (Print each word and your result in "{word} {times}" format).

```python
result = se.flatMap(lambda line: line.split()).countByValue()
for key in result:
    print('{} {}'.format(key, result[key])) 
```

Now, add some normalization to your Spark Job to discart the noise (like strange characters, punctuation, etc...) and avoid repeated words in your results (Print each word and your result in "{word} {times}" format).

```python
from string import punctuation

def clean_line(x):
    for i in punctuation.replace('-',''):
        x = x.replace(i,'')
    return x.lower()

result = se.map(clean_line).flatMap(lambda line: line.split()).countByValue()
for key in result:
    print('{} {}'.format(key, result[key])) 
```

```python
from string import punctuation

def clean_line(x):
    for i in punctuation.replace('-',''):
        x = x.replace(i,'')
    return x.encode('ascii','ignore').decode().lower()

result = se.map(clean_line).flatMap(lambda line: line.split()).countByValue()
for key in result:
    print('{} {}'.format(key, result[key])) 
```

Adjust your Spark Job return the results sorted by count.

```python
se.map(clean_line)\
.flatMap(lambda x: x.split())\
.map(lambda x: (x, 1))\
.reduceByKey(lambda x, y: x + y)\
.sortBy(lambda x: x[1], ascending=False)\
.collect()
```
